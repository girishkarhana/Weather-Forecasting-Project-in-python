```python
import requests
import matplotlib.pyplot as plt
from datetime import datetime

dayin = 3
api_key = "8b2476be20e64e5ca6555052240411"

def get_weather_forecast(city):
    url = f"http://api.weatherapi.com/v1/forecast.json?key={api_key}&q={city}&days={dayin}&aqi=yes&alerts=yes"
    response = requests.get(url)

    if response.status_code == 200:
        data = response.json()

        location = data['location']['name']
        country = data['location']['country']
        forecast_days = data['forecast']['forecastday']

        print(f"\n{dayin}-Day Weather Forecast for {location}, {country}:\n")

        dates = []
        max_temps = []
        min_temps = []
        avg_humidity = []

        for day in forecast_days:
            date = day['date']
            condition = day['day']['condition']['text']
            max_temp = day['day']['maxtemp_c']
            min_temp = day['day']['mintemp_c']
            humidity = day['day']['avghumidity']
            chance_rain = day['day'].get('daily_chance_of_rain', 0)
            chance_snow = day['day'].get('daily_chance_of_snow', 0)
            wind_speed = day['day']['maxwind_kph']

            print(f"Date: {date}")
            print(f"Condition: {condition}")
            print(f"Max Temperature: {max_temp}°C")
            print(f"Min Temperature: {min_temp}°C")
            print(f"Humidity: {humidity}%")
            print(f"Chance of Rain: {chance_rain}%")
            print(f"Chance of Snow: {chance_snow}%")
            print(f"Wind Speed: {wind_speed} km/h")
            print("-" * 30)

            # Add data for visualization
            dates.append(datetime.strptime(date, '%Y-%m-%d'))
            max_temps.append(max_temp)
            min_temps.append(min_temp)
            avg_humidity.append(humidity)

            # Travel Recommendations
            print("\nTravel Recommendations:")
            if chance_rain > 50:
                print("- Pack an umbrella or raincoat!")
            if chance_snow > 50:
                print("- Ensure warm clothing and check road conditions.")
            if wind_speed > 40:
                print("- Be cautious of strong winds, especially when driving.")
            if max_temp > 35:
                print("- Stay hydrated and avoid traveling during peak heat hours.")
            if min_temp < 5:
                print("- Dress warmly and avoid prolonged outdoor exposure.")
            print("\nSuggested Activities:")
            if chance_rain < 30:
                print("- Great day for sightseeing or outdoor activities!")
            else:
                print("- Consider indoor activities like museums or cafes.")
            print("-" * 30)

        # Visualization
        plt.figure(figsize=(10, 5))
        bar_width = 0.4
        date_positions = range(len(dates))

        plt.bar([pos + bar_width / 2 for pos in date_positions], max_temps, width=bar_width, label="Max Temp (°C)", color='tomato')
        plt.bar([pos - bar_width / 2 for pos in date_positions], min_temps, width=bar_width, label="Min Temp (°C)", color='skyblue')

        plt.plot(date_positions, avg_humidity, label="Avg Humidity (%)", marker='o', color='black', linestyle='-', linewidth=2)

        plt.title(f"{dayin}-Day Weather and Travel Forecast for {location}")
        plt.xlabel("Date")
        plt.ylabel("Temperature (°C) and Humidity (%)")
        plt.xticks(ticks=date_positions, labels=[date.strftime("%Y-%m-%d") for date in dates])
        plt.legend()
        plt.grid(axis='y', linestyle='--', alpha=0.7)
        plt.show()
    else:
        print("Failed to retrieve data. Please check the city name or API key.")

city_name = input("Enter city name: ")
get_weather_forecast(city_name)
```
